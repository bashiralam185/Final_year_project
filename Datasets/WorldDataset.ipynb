```python
# importing pands
import pandas as pd


# reading dataset
data2022 = pd.read_csv("all/waqi-covid19-airqualitydata-2022.csv")
data2019_1 = pd.read_csv("all/waqi-covid19-airqualitydata-2019Q1.csv")
data2019_2 = pd.read_csv("all/waqi-covid19-airqualitydata-2019Q2.csv")
data2019_3 = pd.read_csv("all/waqi-covid19-airqualitydata-2019Q3.csv")
data2019_4 = pd.read_csv("all/waqi-covid19-airqualitydata-2019Q4.csv")
data2020_1 = pd.read_csv("all/waqi-covid19-airqualitydata-2020Q1.csv")
data2020_2 = pd.read_csv("all/waqi-covid19-airqualitydata-2020Q2.csv")
data2020_3 = pd.read_csv("all/waqi-covid19-airqualitydata-2020Q3.csv")
data2020_4 = pd.read_csv("all/waqi-covid19-airqualitydata-2020Q4.csv")
data2021_1 = pd.read_csv("all/waqi-covid19-airqualitydata-2021Q1.csv")
data2021_2 = pd.read_csv("all/waqi-covid19-airqualitydata-2021Q2.csv")
data2021_3 = pd.read_csv("all/waqi-covid19-airqualitydata-2021Q3.csv")
data2021_4 = pd.read_csv("all/waqi-covid19-airqualitydata-2021Q4.csv")

```

```python
Bishkek = data2022.query("City == 'Bishkek'")
```

```python
Bishkek.tail()
```

```python
frames = []
def filter_bishkek(df):
    df["Date"] = pd.to_datetime(df["Date"])
    frames.append(df.query("City == 'Bishkek'"))
```

```python
filter_bishkek(data2019_1)
filter_bishkek(data2019_2)
filter_bishkek(data2019_3)
filter_bishkek(data2019_4)
filter_bishkek(data2020_1)
filter_bishkek(data2020_2)
filter_bishkek(data2020_3)
filter_bishkek(data2020_4)
filter_bishkek(data2021_1)
filter_bishkek(data2021_2)
filter_bishkek(data2021_3)
filter_bishkek(data2021_4)
filter_bishkek(data2022)
```

```python
Data = pd.concat(frames)
```

```python
Data = Data.sort_values(by='Date')
```

```python
Data.tail()
```

```python
Data.to_csv("Bishkek_data.csv", index=False)
```
